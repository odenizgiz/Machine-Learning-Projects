# Iris Data

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np 
import pandas as pd 
import seaborn as sns  # visualization
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
def tonumber(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    elif x == 'virginica':
        return 2
    else:
        return 9999
    
iris['species'] = iris['species'].apply(tonumber)
```

## Supervised Learning 

```python
iris_features = iris.drop('species', axis=1)
iris_features.shape
```

```python
iris_target = iris.species
iris_target.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(iris_features, iris_target, random_state=1)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model_nb = GaussianNB()
model_nb.fit(Xtrain, ytrain)
ymodel_nb = model_nb.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, ymodel_nb)
```

```python
Xtest.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

ax[0].scatter(Xtest.petal_length, Xtest.petal_width, c=ymodel_nb, s=50, alpha = 0.5, cmap='rainbow', label='Predicted Result')
ax[0].set_xlabel('Petal Length')
ax[0].set_ylabel('Petal Width')
ax[0].legend(frameon=True, fancybox=True, framealpha=1, shadow=True, loc='upper left');


ax[1].scatter(Xtest.petal_length, Xtest.petal_width, c=ytest, s=50, alpha = .5, cmap='rainbow', label='True Result');
ax[1].set_xlabel('Petal Length')
ax[1].set_ylabel('Petal Width');
ax[1].legend(frameon=True, fancybox=True, framealpha=1, shadow=True, loc='upper left');
```

```python
yprob_nb = model_nb.predict_proba(Xtest)
yprob_nb.round(2)
```

```python
from sklearn.metrics import confusion_matrix
mat_nb = confusion_matrix(ytest, ymodel_nb)
sns.heatmap(mat_nb.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

## Unsupervised Learning

### PCA

```python
iris_features.shape
```

Let's reduce the dimensions from 4 to 2: 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(iris_features)
iris_features_pca = pca.transform(iris_features)
print("original shape:   ", iris_features.shape)
print("transformed shape:", iris_features_pca.shape)
```

Now let's plot the results: 

```python
iris_features_pca[:, 0]
```

```python
# plot with reduced features // not a prediction.

plt.scatter(iris_features_pca[:, 0], iris_features_pca[:, 1], c=iris.species, s=50, 
            alpha=0.3, edgecolor='Black', cmap=plt.cm.get_cmap('viridis', 3))
plt.xlabel('PCA1')
plt.ylabel('PCA2')
# plt.colorbar(ticks=[0, 1, 2], label=['Setosa','Versicolor','Virginica'])
plt.colorbar(ticks=[0, 1, 2]).ax.set_yticklabels(['Setosa','Versicolor','Virginica']);
```

```python
iris_features_pca.shape
```

```python
 iris_target.shape
```

<p> <p/> 

### Clustering

With PCA:

```python
from sklearn.mixture import GaussianMixture 
model_gmm_pca = GaussianMixture(n_components=3, covariance_type='full')
model_gmm_pca.fit(iris_features_pca)                                  # not specifying the label
y_gmm_pca = model_gmm_pca.predict(iris_features_pca)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))


# The predicted result from GMM:
ax[0].scatter(iris_features_pca[:, 0], iris_features_pca[:, 1], c=y_gmm_pca, s=50,    
            alpha=0.3, edgecolor='Black', cmap=plt.cm.get_cmap('viridis', 3), label='Predicted Result')
ax[0].set_xlabel('PCA1')
ax[0].set_ylabel('PCA2')
ax[0].legend(frameon=True, fancybox=True, framealpha=1, shadow=True, loc='upper right');


# The true result from the dataset:
ax[1].scatter(iris_features_pca[:, 0], iris_features_pca[:, 1], c=iris.species, s=50,    
            alpha=0.3, edgecolor='Black', cmap=plt.cm.get_cmap('viridis', 3), label='True result')
ax[1].set_xlabel('PCA1')
ax[1].set_ylabel('PCA2')

ax[1].legend(frameon=True, fancybox=True, framealpha=1, shadow=True, loc='upper right');
```
